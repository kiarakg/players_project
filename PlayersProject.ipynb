```python
# Show number of players and their countries
countries = df.groupby(['nationality'], as_index = False)['short_name'].count()
countries
```

```python
# Show the top 10 countries and their number of players
countries.head(10)
```

```python
# Create a bar plot of top 5 countries and their number of players
countries_top5 = countries.head(5)
countries_top5
```

```python
plt.figure(figsize = (10,6))
plt.bar(countries_top5['nationality'], countries_top5['short_name'], color = "green")
plt.xlabel('Nationality', size = 15)
plt.ylabel('Number of Players', size = 15)
plt.title('Players per Country')
# plt.savefig("num_of_each_species_bchart.png")
plt.show()
```
