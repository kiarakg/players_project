```python
# Show number of players and their countries
countries = df.groupby(['nationality'], as_index = False)['short_name'].count()
countries
```

```python
# Show the top 10 countries and their number of players
countries.head(10)
```

```python
# Create a bar plot of top 5 countries and their number of players
countries_top5 = countries.head(5)
countries_top5
```

```python
plt.figure(figsize = (10,6))
plt.bar(countries_top5['nationality'], countries_top5['short_name'], color = "green")
plt.xlabel('Nationality', size = 15)
plt.ylabel('Number of Players', size = 15)
plt.title('Players per Country')
# plt.savefig("top5_countries_bchart.png")
plt.show()
```

```python
# Show top 5 players short name and wages
top5_players = df.head(5)
top5_players
```

```python
# Show top 5 players short name and wages
top_wages = top5_players.groupby(['short_name','wage_eur'])['long_name'].count()
top_wages
```

```python
sorted_df = df.sort_values("short_name")
sorted_df
```

```python
# Show top 5 players short name and wages that are getting highest salaries
top_wages = df[['short_name','wage_eur']].sort_values(by = 'wage_eur', ascending = False).head(5)
top_wages
```

```python
# Create a bar plot of top 5 players short name and wages that are getting highest salaries
plt.figure(figsize = (10,6))
plt.bar(top_wages['short_name'], top_wages['wage_eur'], color = "purple")
plt.xlabel('Players', size = 15)
plt.ylabel('Wages', size = 15)
plt.title('Highest Salaries')
# plt.savefig("top5_countries_bchart.png")
plt.show()
```
